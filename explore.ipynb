```python
# import dependencies
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Loading the Datasets in Chunks

```python
# load the dataset
store = []
chunk_size = 500000

for chunk in pd.read_json('arxiv-metadata-oai-snapshot.json', lines = True, chunksize = chunk_size):
    store.append(chunk)

data = pd.concat(store, ignore_index = True)
```

#### download the pdfs

```python
base_url = 'https://arxiv.org/pdf'

# generate the pdf links
def generate_pdf_links(row):
    return f"{base_url}/{row['id']}.pdf"

# download paths
data_econGN_2020['pdf_link'] = data_econGN_2020.apply(generate_pdf_links, axis=1)
```

```python
data_econGN_2020.head()
```

```python
import requests
import time
```

```python
def download_pdf(pdf_link, save_path):
    response = requests.get(pdf_link)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            file.write(response.content)
    else:
        print(f"Failed to download {pdf_link}")

# download the pdfs

save_dir = '/Users/ikedijacobs/Downloads/arxiv pdfs/'

for index, row in data_econGN_2020.iterrows():
    pdf_link = row['pdf_link']
    save_path = f"{save_dir}{row['id']}.pdf"
    download_pdf(pdf_link, save_path)
    time.sleep(1)
```

```python
print("Download complete")
```
